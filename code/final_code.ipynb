```python
def compute_lcc(G):
    '''takes a graph and returns the size of the largest connected component'''
    if G.number_of_nodes() == 0:
        return 0
    else:
        max_comp= max(nx.connected_components(G), key=len)
        return len(max_comp)
    
def compute_i(G, i):
    '''takes a graph and the size of the initial largest connected component
    returns teh current size of the largest connected component divided by the initial size (i index)
    '''
    if G.number_of_nodes() == 0:
        return 0
    else:
        max_comp= max(nx.connected_components(G), key=len)
        return len(max_comp)/i

def compute_r(g, sorted_nodes):
    '''
    computes the robustness measure following this formula:
    R = SUM(LCCi)/N where LCCi is the relative size of the largest connected component after the removal of i nodes

    this computation is only calculated for sequential removals
    '''
    G=g.copy()
    SUM=0
    N= G.number_of_nodes()

    if type(sorted_nodes[0])==tuple:

        for i in range(N):

            n=sorted_nodes[i][0]
            G.remove_node(n)
            lcc= compute_lcc(G)/N
            SUM+=lcc
    else:
        for i in range(N):

            n=sorted_nodes[i]
            G.remove_node(n)
            lcc= compute_lcc(G)/N
            SUM+=lcc

    return SUM/N

# def compute_r_50_batches(attack):
#     '''takes a tuple of dictionaries (output of attack) that contains the lcc of attack'''
#     lcc_attack=attack[3]
#     r = sum(list(lcc_attack.values()))/len(list(lcc_attack.values()))
#     return r

# def compute_r_100_batches(g, sorted_nodes):
#     G=g.copy()
#     N=G.number_of_nodes()
#     batch_size = N//100
#     SUM=0
#     if type(sorted_nodes[0])==tuple:

#         for i in range(50):
#             for j in range(batch_size):
#                 n = sorted_nodes[i*batch_size+j][0]
#                 G.remove_node(n)
#             lcc = compute_lcc(G)/N
#             SUM+=lcc
    
#     else:
#         for i in range(50):
#             for j in range(batch_size):
#                 n = sorted_nodes[i*batch_size+j]
#                 G.remove_node(n)
#             lcc = compute_lcc(G)/N
#             SUM+=lcc

#     return SUM/100


def random_attack(g):
    '''randomly attacks a graph and returns:
    list of LCCs
    list of I indices
    list of global efficiencies
    time of execution

    its done per batches - can not calculate the robustness measure internally
    '''
    G=g.copy()
    nodes = list(G.nodes())
    random.shuffle(nodes)
    #divide the nodes into 100 batches
    fix=G.number_of_nodes()
    batch_size = fix//100
    LCC={}
    Inter={}
    E={}
    lcc_i = max([len(c) for c in nx.connected_components(G)])
    t=time.time()

    for i in range(50):
        for j in range(batch_size):
            n = nodes[i*batch_size+j]
            G.remove_node(n)
        lcc = compute_lcc(G)
        i_index= compute_i(G, lcc_i)
        LCC[i] = lcc/fix
        Inter[i]=i_index
        E[i]=nx.algorithms.global_efficiency(G)
    e=time.time()
    t=e-t
    return LCC, Inter, E, t

def random_attack_edge(g):
    '''apply random attack on edges:
    returns:
    list of LCCs
    list of I indices
    list of global efficiencies
    time of execution
    '''
    G = g.copy()
    edges = list(G.edges())
    random.shuffle(edges)
    fix=G.number_of_edges()
    batch_size = fix//100
    LCC={}
    Inter={}
    E={}
    lcc_i = max([len(c) for c in nx.connected_components(G)])
    t=time.time()

    for i in range(50):
        for j in range(batch_size):
            e = edges[i*batch_size+j]
            G.remove_edge(e[0], e[1])
        lcc = compute_lcc(G)
        i_index= compute_i(G, lcc_i)
        LCC[i] = lcc/fix
        Inter[i]=i_index
        E[i]=nx.algorithms.global_efficiency(G)

    e=time.time()
    t=e-t
    return LCC, Inter, E, t

def partition_list(input_list, num=100):
    '''takes a list and splits it into num partitions'''''
    n = len(input_list) // num
    remainder = len(input_list) % num
    partitions = []
    for i in range(num):
        start = i * n + min(i, remainder)
        end = start + n + (i < remainder)
        partitions.append(input_list[start:end])
    return partitions

# def attack(g, sorted_nodes, batch_num=100):
#     '''
#     This funcion takes a graph and a list of nodes sorted by their centrality measure in order to perform batch attacks on the graph
#     It splits the list of nodes into 100 batches (by default) and removes the nodes in each batch from the graph
#     (uses the helper function partition_list) 

#     praram:
#         g: graph
#         sorted_nodes: list of nodes sorted by their centrality measure
#         batch_num: number of batches to split the list of nodes into
#     returns: a tuple of 6 items
#         [0] norm_LCC: dictionary of normalized LCCs
#         [1] Inter: dictionary of I indices
#         [2] E: dictionary of global efficiencies
#         [3] R: float robustness measure
#         [4] InterE: float global efficiency
#         [5] t: float time of execution
#     '''
#     G=g.copy()  
#     N=G.number_of_nodes()
#     pi = partition_list(sorted_nodes, batch_num)
#     # print(len(pi))

#     lcc=0;norm_lcc=0;i_index=0;e=0
#     norm_LCC={};LCC={};Inter={};E={}
#     R=[]; InterE=[]

#     t=time.time()

#     for batch in range((len(pi))): #looping through each batch
#         for node in pi[batch]:
#             G.remove_node(node)
#             lcc = compute_lcc(G)
#             i_index= compute_i(G, N)
#             norm_lcc = lcc/N
#             e = nx.algorithms.global_efficiency(G)

#             R.append(norm_lcc)
#             InterE.append(e)

#         norm_LCC[batch]=norm_lcc
#         LCC[batch]=lcc
#         Inter[batch]=i_index
#         E[batch]=e

#     e=time.time()
#     t=e-t

#     #R at halftime doesn not hold according to formula
#     R=sum(R)/len(R)
#     InterE=sum(InterE)/len(InterE)

#     return norm_LCC, Inter, E, R, InterE, t

# def random_attack(g, batch_num=100):
#     '''
#     This funcion takes a graph in order to perform batch attacks randomly on the graph
#     It splits the list of nodes into 100 batches (by default) and removes the nodes in each batch from the graph
#     WILL USE THE NORMAL ATTACK METHOD BUT GIVE IT RANDOMLY SHUFFLED LIST OF NODES INSTEAD OF SORTED ONES

#     praram:
#         g: graph
#         sorted_nodes: list of nodes sorted by their centrality measure
#         batch_num: number of batches to split the list of nodes into
#     returns: a tuple of 6 items
#         [0] norm_LCC: dictionary of normalized LCCs
#         [1] Inter: dictionary of I indices
#         [2] E: dictionary of global efficiencies
#         [3] R: float robustness measure
#         [4] InterE: float global efficiency
#         [5] t: float time of execution
#     '''
#     G=g.copy()
#     nodes = list(G.nodes())
#     shuffled_nn = random.shuffle(nodes)
#     return attack(G, shuffled_nn, batch_num)

def attack(g, sorted_nodes):
    G=g.copy()
    fix=G.number_of_nodes()
    batch_size = fix//100
    norm_LCC={}
    LCC={}
    Inter={}
    E={}
    lcc_i = max([len(c) for c in nx.connected_components(G)])

    t=time.time()
    for i in range(50):
        for j in range(batch_size):
            n = sorted_nodes[i*batch_size+j]
            G.remove_node(n)
        lcc = compute_lcc(G)
        i_index= compute_i(G, lcc_i)
        norm_LCC[i] = lcc/fix
        LCC[i]=lcc
        Inter[i]=i_index
        E[i]=nx.algorithms.global_efficiency(G)
    e=time.time()

    t=e-t
    return norm_LCC, Inter, E, t

def greedy_list(g) ->list:
    '''takes a graph and computes a list of node that follows the greedy attack scheme
    Get an initial list of best nodes to attack - network of n nodes, try n times:
        Copy the network in a new test network
        Remove a node from the test
        Assess robustness score after its removal
    Do this for the n nodes, sort them by order of most impactful attacks and emply the sequential 
    '''
    n=g.number_of_nodes()
    nodes=list(g.nodes())
    stat={}
    for i in range(n):
        G=g.copy()
        G.remove_node(nodes[i])
        lcc=compute_lcc(G)
        stat[nodes[i]]=lcc
    #sort the nodes keys in ascending order of their values in the dictionary
    nodes_sorted =  sorted(stat.items(), key=lambda x: x[1], reverse=False) 
    #get only the keys 
    nodes_sorted_list = [x[0] for x in nodes_sorted]
    return nodes_sorted_list

def get_MEB(G:nx.Graph, k:int, bet:dict)->float:
    '''
    takes a graph and a node k, returns the MEB centrality (dictionary) following this formula:
    MEBk= BCk * SUM(log(BCi)) for all i neighbor of k
    '''
    BCk = bet[k][1]
    neighbors = list(G.neighbors(k))
    BC_neighbors = [bet[x][1] for x in neighbors]
    sum_log_BC_neighbors = sum([math.log(x+1e-06) for x in BC_neighbors]) #if betweeness is 0, consider log(0)=-50
    MEBk = - BCk * sum_log_BC_neighbors
    return MEBk

def MEB_centrality(G:nx.Graph, bet)->dict:
    '''takes a graph and returns a dictionary of MEB centrality for each node'''
    MEB = {}
    for node in G.nodes():
        MEB[node] = get_MEB(G, node, bet)
    return MEB

def MEB_centrality_com(G):
    MEB={}
    for node in G.nodes():
        MEB[node] = get_MEB_com(G, node)
    return MEB

def get_MEB_com(G:nx.Graph, k:int)->float:
    '''
    takes a graph and a node k, returns the MEB centrality following this formula:
    MEBk= - BCk * SUM(log(BCi)) for all i neighbor of k
    '''
    bet=nx.betweenness_centrality(G)
    BCk = bet[k]
    neighbors = list(G.neighbors(k))
    BC_neighbors = [bet[x] for x in neighbors]
    sum_log_BC_neighbors = sum([math.log(x+1e-06) for x in BC_neighbors]) #if betweeness is 0, consider log(0)=-50
    MEBk = - BCk * sum_log_BC_neighbors
    return MEBk

def MEC(G:nx.Graph, k:int, clos:dict)->float:
    '''
    takes a graph and a node k, returns the MEC centrality following this formula:
    MECk= - CCk * SUM(log(CCi)) for all i neighbor of k
    '''
    CCk = clos[k][1]
    neighbors = list(G.neighbors(k))
    CC_neighbors = [clos[x][1] for x in neighbors]
    sum_log_CC_neighbors = sum([math.log(x) for x in CC_neighbors]) 
    MECk = - CCk * sum_log_CC_neighbors
    return MECk

def MEC_centrality(G:nx.Graph, clos)->dict:
    '''takes a graph and returns a dictionary of MEC centrality for each node'''
    MEC_dict = {}
    for node in G.nodes():
        MEC_dict[node] = MEC(G, node, clos)
    return MEC_dict

def get_MEC_com(G:nx.Graph, k:int)->float:
    '''
    takes a graph and a node k, returns the MEC centrality following this formula:
    MECk= - CCk * SUM(log(CCi)) for all i neighbor of k
    '''
    clos=nx.closeness_centrality(G)
    CCk = clos[k]
    neighbors = list(G.neighbors(k))
    CC_neighbors = [clos[x] for x in neighbors]
    sum_log_CC_neighbors = sum([math.log(x) for x in CC_neighbors]) 
    MECk = - CCk * sum_log_CC_neighbors
    return MECk

def MEC_centrality_com(G):
    MEC={}
    for node in G.nodes():
        MEC[node] = get_MEC_com(G, node)
    return MEC

def MED_centrality(G, deg):
    MED_dict = {}
    for node in G.nodes():
        MED_dict[node] = MED(G, node, deg)
    return MED_dict

def MED(G:nx.Graph, k:int, deg:dict)->float:
    '''
    takes a graph and a node k, returns the MED centrality following this formula:
    MEDk= - DEGk * SUM(log(DEGi)) for all i neighbor of k
    '''
    DEGk = deg[k][1]
    neighbors = list(G.neighbors(k))
    DEG_neighbors = [deg[x][1] for x in neighbors]
    sum_log_DEG_neighbors = sum([math.log(x) for x in DEG_neighbors]) #if betweeness is 0, consider log(0)=-50
    MEDk = - DEGk * sum_log_DEG_neighbors
    return MEDk

def get_MED_com(G:nx.Graph, k:int)->float:
    '''
    takes a graph and a node k, returns the MED centrality following this formula:
    MEDk= - DEGk * SUM(log(DEGi)) for all i neighbor of k
    '''
    deg=nx.degree_centrality(G)
    DEGk = deg[k]
    neighbors = list(G.neighbors(k))
    DEG_neighbors = [deg[x] for x in neighbors]
    sum_log_DEG_neighbors = sum([math.log(x) for x in DEG_neighbors]) #if betweeness is 0, consider log(0)=-50
    MEDk = - DEGk * sum_log_DEG_neighbors
    return MEDk

def MED_centrality_com(G):
    MED={}
    for node in G.nodes():
        MED[node] = get_MED_com(G, node)
    return MED


def proximity(node, shortest_paths):
    '''
    takes a graph and a node, returns the proximity centrality of that node
    '''
    sum_of_shortest_paths = sum(shortest_paths[node])
    # print(sum_of_shortest_paths)
    sum_of_all_pairs_shortest_paths = sum(sum(shortest_paths))
    proximity = sum_of_all_pairs_shortest_paths / sum_of_shortest_paths 
    return proximity

def proximity_centrality(G:nx.Graph)->dict:
    '''takes a graph and returns a dictionary of proximity centrality for each node'''
    nodes = list(G.nodes())
    shortest_paths = nx.floyd_warshall_numpy(G)
    pr = {}
    for i, node in enumerate(nodes):
        pr[node] = proximity(i, shortest_paths)
    return pr

```

##### DWT

```python
def compute_S(G:nx.Graph, i:int, j:int):
    '''takes node  and j and computes Sij'''
    Ni = G.neighbors(i)
    Nj = G.neighbors(j)

    intersection = set(Ni).intersection(set(Nj))

    if len(intersection)==0: #if inter is empty
        S=0
    else:
        ki= G.degree(i)
        kj= G.degree(j)
        S= len(intersection)/math.sqrt(ki*kj)

    return S

def DWT_i(G, i):
    '''takes a graph, a node and computes the DWT_i'''
    ki= G.degree(i)
    Ni= G.neighbors(i)
    DWT=0


    for j in Ni: #loop thru all neighbors and get the Sij
        Sij=compute_S(G, i, j)
        DWT += ( (1+Sij)/ki ) ** 2

    
    return DWT

def DWT_centrality(G):
    # Calculate the DWT centrality for each node in the graph
    centrality = {}
    for node in G.nodes():
        centrality[node] = DWT_i(G, node)
    return centrality

dwt_centrality_arenas_email = sorted(DWT_centrality(arenas_email).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_arenas_email_nodes =[x[0] for x in dwt_centrality_arenas_email]
dwt_attack_arenas_email = attack(arenas_email, dwt_centrality_arenas_email_nodes)

dwt_centrality_bn_cat_mixed_species_brain_1 = sorted(DWT_centrality(bn_cat_mixed_species_brain_1).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_bn_cat_mixed_species_brain_1_nodes =[x[0] for x in dwt_centrality_bn_cat_mixed_species_brain_1]
dwt_attack_bn_cat_mixed_species_brain_1 = attack(bn_cat_mixed_species_brain_1, dwt_centrality_bn_cat_mixed_species_brain_1_nodes)

dwt_centrality_bn_macaque_rhesus_brain_2 = sorted(DWT_centrality(bn_macaque_rhesus_brain_2).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_bn_macaque_rhesus_brain_2_nodes =[x[0] for x in dwt_centrality_bn_macaque_rhesus_brain_2]
dwt_attack_bn_macaque_rhesus_brain_2 = attack(bn_macaque_rhesus_brain_2, dwt_centrality_bn_macaque_rhesus_brain_2_nodes)

dwt_centrality_circuits_s208 = sorted(DWT_centrality(circuits_s208).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_circuits_s208_nodes =[x[0] for x in dwt_centrality_circuits_s208]
dwt_attack_circuits_s208 = attack(circuits_s208, dwt_centrality_circuits_s208_nodes)

dwt_centrality_circuits_s420 = sorted(DWT_centrality(circuits_s420).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_circuits_s420_nodes =[x[0] for x in dwt_centrality_circuits_s420]
dwt_attack_circuits_s420 = attack(circuits_s420, dwt_centrality_circuits_s420_nodes)

dwt_centrality_circuits_s838 = sorted(DWT_centrality(circuits_s838).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_circuits_s838_nodes =[x[0] for x in dwt_centrality_circuits_s838]
dwt_attack_circuits_s838 = attack(circuits_s838, dwt_centrality_circuits_s838_nodes)

# dwt_centrality_dolphins = sorted(DWT_centrality(dolphins).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_dolphins_nodes =[x[0] for x in dwt_centrality_dolphins]
# dwt_attack_dolphins = attack(dolphins, dwt_centrality_dolphins_nodes)

dwt_centrality_e_coli = sorted(DWT_centrality(e_coli).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_e_coli_nodes =[x[0] for x in dwt_centrality_e_coli]
dwt_attack_e_coli = attack(e_coli, dwt_centrality_e_coli_nodes)

dwt_centrality_facebook_0 = sorted(DWT_centrality(facebook_0).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_facebook_0_nodes =[x[0] for x in dwt_centrality_facebook_0]
dwt_attack_facebook_0 = attack(facebook_0, dwt_centrality_facebook_0_nodes)

dwt_centrality_facebook_107 = sorted(DWT_centrality(facebook_107).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_facebook_107_nodes =[x[0] for x in dwt_centrality_facebook_107]
dwt_attack_facebook_107 = attack(facebook_107, dwt_centrality_facebook_107_nodes)

dwt_centrality_facebook_1684 = sorted(DWT_centrality(facebook_1684).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_facebook_1684_nodes =[x[0] for x in dwt_centrality_facebook_1684]
dwt_attack_facebook_1684 = attack(facebook_1684, dwt_centrality_facebook_1684_nodes)

dwt_centrality_facebook_348 = sorted(DWT_centrality(facebook_348).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_facebook_348_nodes =[x[0] for x in dwt_centrality_facebook_348]
dwt_attack_facebook_348 = attack(facebook_348, dwt_centrality_facebook_348_nodes)

dwt_centrality_facebook_414 = sorted(DWT_centrality(facebook_414).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_facebook_414_nodes =[x[0] for x in dwt_centrality_facebook_414]  
dwt_attack_facebook_414 = attack(facebook_414, dwt_centrality_facebook_414_nodes)

dwt_centrality_facebook_686 = sorted(DWT_centrality(facebook_686).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_facebook_686_nodes =[x[0] for x in dwt_centrality_facebook_686]
dwt_attack_facebook_686 = attack(facebook_686, dwt_centrality_facebook_686_nodes)

# dwt_centrality_fb_pages_food = sorted(DWT_centrality(fb_pages_food).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_fb_pages_food_nodes =[x[0] for x in dwt_centrality_fb_pages_food]
# dwt_attack_fb_pages_food = attack(fb_pages_food, dwt_centrality_fb_pages_food_nodes)

# dwt_centrality_karate = sorted(DWT_centrality(karate).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_karate_nodes =[x[0] for x in dwt_centrality_karate]
# dwt_attack_karate = attack(karate, dwt_centrality_karate_nodes)

dwt_centrality_polbooks = sorted(DWT_centrality(polbooks).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_polbooks_nodes =[x[0] for x in dwt_centrality_polbooks]
dwt_attack_polbooks = attack(polbooks, dwt_centrality_polbooks_nodes)

# dwt_centrality_soc_firm_hi_tech = sorted(DWT_centrality(soc_firm_hi_tech).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_soc_firm_hi_tech_nodes =[x[0] for x in dwt_centrality_soc_firm_hi_tech]
# dwt_attack_soc_firm_hi_tech = attack(soc_firm_hi_tech, dwt_centrality_soc_firm_hi_tech_nodes)

# dwt_centrality_soc_tribes = sorted(DWT_centrality(soc_tribes).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_soc_tribes_nodes =[x[0] for x in dwt_centrality_soc_tribes]
# dwt_attack_soc_tribes = attack(soc_tribes, dwt_centrality_soc_tribes_nodes)

dwt_centrality_word_adjacencies = sorted(DWT_centrality(word_adjacencies).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_word_adjacencies_nodes =[x[0] for x in dwt_centrality_word_adjacencies]
dwt_attack_word_adjacencies = attack(word_adjacencies, dwt_centrality_word_adjacencies_nodes)

# dwt_centrality_net911 = sorted(DWT_centrality(net_911).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_net911_nodes =[x[0] for x in dwt_centrality_net911]
# dwt_attack_net911 = attack(net_911, dwt_centrality_net911_nodes)

dwt_centrality_corruption = sorted(DWT_centrality(corruption).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_corruption_nodes =[x[0] for x in dwt_centrality_corruption]
dwt_attack_corruption = attack(corruption, dwt_centrality_corruption_nodes)

dwt_centrality_crime_net = sorted(DWT_centrality(crime_net).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_crime_net_nodes =[x[0] for x in dwt_centrality_crime_net]
dwt_attack_crime_net = attack(crime_net, dwt_centrality_crime_net_nodes)

# dwt_centrality_digg = sorted(DWT_centrality(digg).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_digg_nodes =[x[0] for x in dwt_centrality_digg]
# dwt_attack_digg = attack(digg, dwt_centrality_digg_nodes)

dwt_centrality_jazz = sorted(DWT_centrality(jazz).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_jazz_nodes =[x[0] for x in dwt_centrality_jazz]
dwt_attack_jazz = attack(jazz, dwt_centrality_jazz_nodes)

dwt_centrality_petster_hamster = sorted(DWT_centrality(petster_hamster).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_petster_hamster_nodes =[x[0] for x in dwt_centrality_petster_hamster]
dwt_attack_petster_hamster = attack(petster_hamster, dwt_centrality_petster_hamster_nodes)

# dwt_centrality_router = sorted(DWT_centrality(router).items(), key=lambda x: x[1], reverse=True)
# dwt_centrality_router_nodes =[x[0] for x in dwt_centrality_router]
# dwt_attack_router = attack(router, dwt_centrality_router_nodes)

dwt_centrality_email = sorted(DWT_centrality(email).items(), key=lambda x: x[1], reverse=True)
dwt_centrality_email_nodes =[x[0] for x in dwt_centrality_email]
dwt_attack_email = attack(email, dwt_centrality_email_nodes)

```

```python
def split_community(g:nx.Graph):
    '''
    takes a graph and tries 10 louvain community detections and returns the split with the highest modularity
    '''
    best_modularity = 0
    best_partition = None
    for i in range(10):
        partition = nx.community.louvain_communities(g, seed=random.randint(0, 1000))
        modularity = nx.community.modularity(g, partition)
        if modularity > best_modularity:
            best_modularity = modularity
            best_partition = partition
    return best_partition

def get_boundary_nodes(G, partition):   
    '''
    takes:
        graph and partiton - list of sets of nodes - as input
    
    returns:
        the boundary nodes in a list
    '''
    boundary_edges = set()
    for i in range(len(partition)):
        for j in range(len(partition)):
            if i != j:
                boundary_edges.update(nx.edge_boundary(G, partition[i], partition[j]))

    b_nodes = []
    for i in range(len(boundary_edges)):
        b_nodes.append(list(boundary_edges)[i][0])
        b_nodes.append(list(boundary_edges)[i][1])
    b_nodes = list(set(b_nodes))
    return b_nodes

def _community_attack(g:nx.Graph, partition, centrality_func_bound, centrality_func_internal):
    '''
    takes:
    a graph, a partition, a list of nodes sorted based on their centrality, and a centrality function to recalculate centralities after removal of boundary nodes
    
    returns: 
    LCC, I inex, efficiency, R, InterE, time taken

    ! it requires a helper method: get_boundary_nodes(g, pi)
    ! this itself is a helper method for community_attack(g, centrality_func_bound, centrality_func_internal)
    '''
    G=g.copy(); N=G.number_of_nodes()
    boundary_nodes = get_boundary_nodes(g, partition) ;B=len(boundary_nodes)
    LCC={};Inter={}; Eff={}
    R=[]; InterE=[]

    boundary_sorted = [x[0] for x in sorted(centrality_func_bound(G).items(), key=lambda x: x[1], reverse=True) if x[0] in boundary_nodes]

    t=time.time()
    for i in range(B):
        b_node = boundary_sorted[i]
        G.remove_node(b_node)
        lcc=compute_lcc(G)/N; i_index=compute_i(G, N); eff=nx.algorithms.global_efficiency(G)
        R.append(lcc); InterE.append(eff)
        LCC[i]=lcc; Inter[i]=i_index; Eff[i]=eff

    internal_sorted= [x[0] for x in sorted(centrality_func_internal(G).items(), key=lambda x: x[1], reverse=True) if x[0] not in boundary_nodes]
    for i in range(N-B):
        node = internal_sorted[i]
        G.remove_node(node)
        lcc=compute_lcc(G)/N; i_index=compute_i(G, N); eff=nx.algorithms.global_efficiency(G)
        R.append(lcc); InterE.append(eff)
        LCC[B+i]=lcc; Inter[B+i]=i_index; Eff[B+i]=eff
    
    e=time.time()
    t=e-t
    R=sum(R)/len(R); InterE=sum(InterE)/len(InterE)
    return LCC, Inter, Eff, R, InterE, t, B #returns the number of boundary nodes to highligh them in the plot

def community_attack(g:nx.Graph, centrality_func_bound, centrality_func_internal):
    '''
    Takes:
        a graph, a centrality function to calculate centralities of boundary nodes, and a centrality function to calculate centralities of internal nodes
    Returns:
        LCC, I inex, efficiency, R, InterE, time taken
    
    Runs 10 louvain community detections and uses the split with the highest modularity
    '''
    p = split_community(g)
    return _community_attack(g, p, centrality_func_bound, centrality_func_internal)

def seq_attack(g:nx.Graph, centrality_func):
    '''
    takes a graph and a centrality function and returns the LCC, I inex, efficiency, R, InterE, time taken
    recalculates centrality after each removal
    '''
    G=g.copy(); N=G.number_of_nodes()
    LCC={};Inter={}; Eff={}
    R=[]; InterE=[]

    t=time.time()
    for i in range(N):
        sorted_nodes=[x[0] for x in sorted(centrality_func(G).items(), key=lambda x: x[1], reverse=True)]
        node = sorted_nodes[0]
        G.remove_node(node)
        lcc=compute_lcc(G)/N; i_index=compute_i(G, N); eff=nx.algorithms.global_efficiency(G)
        R.append(lcc); InterE.append(eff)
        LCC[i]=lcc; Inter[i]=i_index; Eff[i]=eff
    
    e=time.time()
    t=e-t
    R=sum(R)/len(R); InterE=sum(InterE)/len(InterE)
    return LCC, Inter, Eff, R, InterE, t
```
